 <center><h1 style="color:#C3B091;"> 1.3 Pre-Processing üßπ</center>

```python
import string
string.punctuation
```

```python
train.head()
```

```python

# Download stopwords if not already downloaded
nltk.download('stopwords')
```

```python
import string
def cleaning(content):
    punc=[char for char in content if char not in string.punctuation]
    punc_join="".join(punc)
    punc_join_word= [word for word in punc_join.split() if word.lower() not in stopwords.words('english')]
    return punc_join_word

clean_text= train['text'].apply(cleaning)
```

```python
clean_text
```

```python
vectorizer=CountVectorizer(analyzer=cleaning,dtype=np.uint8)
```

```python
text_vector=vectorizer.fit_transform(train['text'])
```

```python
print(vectorizer.get_feature_names_out())
#print(vectorizer.get_feature_names())
```

```python
print(text_vector.toarray())
```

```python
text_vector.shape
```

```python
X=pd.DataFrame(text_vector.toarray())
```

```python
y=train['polarity']
```

---

 <center><h1 style="color:#C3B091;"> 1.4 Model üñ•Ô∏è</center>

```python
X.shape, y.shape
```

```python
X_train, X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
NB_classifier=MultinomialNB()
NB_classifier.fit(X_train,y_train)
```

```python
y_pred=NB_classifier.predict(X_test)
```

```python
cm=confusion_matrix(y_test,y_pred)
```

```python
sns.heatmap(cm,annot=True)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
```
